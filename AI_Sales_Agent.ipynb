```python
from datetime import datetime

leads = [
    {"Name": "Ravi Kumar", "Company": "TCS", "Last Contact": "2026-03-20"},
    {"Name": "Priya Sharma", "Company": "Infosys", "Last Contact": "2026-03-26"}
]

def generate_email(name, company):
    return f"Hi {name}, I’d love to discuss how our solution can help {company} improve sales."

def detect_risk(last_contact):
    last_date = datetime.strptime(last_contact, "%Y-%m-%d")
    days = (datetime.now() - last_date).days
    return "HIGH RISK" if days > 3 else "SAFE"

def generate_followup(name):
    return f"Hi {name}, just following up on my previous email. Looking forward to your response."

for lead in leads:
    print(f"\n===== {lead['Name']} ({lead['Company']}) =====")
    
    print("\n📧 Generated Email:\n", generate_email(lead["Name"], lead["Company"]))
    
    risk = detect_risk(lead["Last Contact"])
    print("\n⚠️ Risk Status:", risk)
    
    if risk == "HIGH RISK":
        print("\n🔁 Follow-up Email:\n", generate_followup(lead["Name"]))
```
